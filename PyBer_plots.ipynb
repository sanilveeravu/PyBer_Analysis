```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
#plt.figure(figsize=(10,8))
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot.
#plt.plot(x_axis,y_axis)
#plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Object Oriented Approach

#fig, ax = plt.subplots(figure=(w,h))
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis,label="Boston",color="green", marker="D", linewidth=2)
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.grid()
ax.set_ylim(0,45)
```

```python
#Bar grap
plt.bar(x_axis,y_axis,color='green',label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
```
